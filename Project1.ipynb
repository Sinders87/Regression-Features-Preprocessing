```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

X = df.drop(columns=['DEATH_EVENT'])
y = df['DEATH_EVENT']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

log_regression = LogisticRegression(random_state=42)
log_regression.fit(X_train_scaled, y_train)
log_regression_preds = log_regression.predict(X_test_scaled)

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)

log_regression_cm = confusion_matrix(y_test, log_regression_preds)
rf_cm = confusion_matrix(y_test, rf_preds)
log_regression_cv_scores = cross_val_score(log_regression, scaler.fit_transform(X), y, cv=5)
rf_cv_scores = cross_val_score(rf, X, y, cv=5)

print("Logistic Regression Classification Report: ")
print(classification_report(y_test, log_regression_preds))
print("Random Forest Classification Report: ")
print(classification_report(y_test, rf_preds))

print("Cross-Validation Scores:")
print(f"Logistic Regression - Mean Accuracy: {log_regression_cv_scores.mean():.2f}, Std Dev: {log_regression_cv_scores.std():.2f}")
print(f"Random Forest       - Mean Accuracy: {rf_cv_scores.mean():.2f}, Std Dev: {rf_cv_scores.std():.2f}")
```
